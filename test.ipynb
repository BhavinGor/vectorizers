```python
from vect_utils import *
import pandas as pd

# Sample DataFrame
data = {
    'text': [
        'The quick brown fox jumps over the lazy dog.',
        'Never gonna give you up, never gonna let you down.',
        'To be or not to be, that is the question.'
    ],
    'text_1': [
        'The quick brown fox jumps over the lazy dog.',
        'Never gonna give you up, never gonna let you down.',
        'To be or not to be, that is the question.'
    ],
    'numeric_column': [42, 17, 23],
    'list_column': [['apple', 'banana', 'cherry'], ['cat', 'dog'], ['sun', 'moon']]
}

df = pd.DataFrame(data)

```

```python
df
```

```python
import pickle
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Example ColumnTransformer
column_transformer = ColumnTransformer(
    transformers=[
        ('bm25', BM25Transformer(max_features=5), 'text'),
        ('bm42', BM42Transformer(max_features=3), 'text_1'),
        ('mlb', MultiLabelBinarizerTransformer(), 'list_column'),
        ('onehot', OneHotEncoder(), ['numeric_column']),


    ],
    remainder='passthrough'
)

column_transformer.fit_transform(df)

# Save the fitted ColumnTransformer using pickle
with open('column_transformer.pkl', 'wb') as f:
    pickle.dump(column_transformer, f)
```

```python
# # Load the ColumnTransformer
with open('column_transformer.pkl', 'rb') as f:
    loaded_column_transformer = pickle.load(f)

loaded_column_transformer
```

```python
loaded_column_transformer.transform(df)
```
